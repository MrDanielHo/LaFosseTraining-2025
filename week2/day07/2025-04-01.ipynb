### How to make Moka Pot coffee
* ground 15g of medium ground coffee per cup
* Unscrew the moka pot and seperate the chambers
* Fill the water chamber full with cold water up to 3/4 full.
* Place on the counter top
* Place the funnel filter on top of the water chamber
* place the reducer disk on top of the funnel filter
* place filter on top of the reducer disk
* place the coffee chamber on top and turn clockwise until locked
* Place the moka pot onto a gas/electric hob on a medium heat
* listen to the audio signal from the moka pot
* turn off the source of heat
* Open the lid to the moka pot and wait for the liquid to pass through the top.
* (optional) take the moka pot and run cold over the bottom chamber to cool it down.
* Pour the coffee into a cup/carafe

```js
function anagrammar (str1, str2) {
    function strTransformation (str3) {
        return str3.replaceAll(' ', '').toLowerCase()
    }

    let str1nosp = strTransformation(str1).sort()
    let str2nosp = strTransformation(str2).sort()

    if (str1nosp == str2nosp) {
        return `${str1} is an anagram of ${str2}`
    } else {
        return `${str1} is not an anagram of ${str2}`

}
```
