## Case Study - Futurehomes
Founded 7 years ago, we are a mid-size real estate company that has traditionally focussed on sales. We have had a successful start, however have noticed a fall in property sales over the past 12 months. Having seen the trends outlined below we are looking for ways to extend our offering and are beginning to look a the rental market. 
 
We are looking to identify pain points in the "finding/keeping a good roommate" journey and to find ways of solving those pain points.
 
A few trends of note: 54% of the world's population live in urban communities, the average marriage age for men is 29 (up from 26 two decades ago) and for women is 27 (up from 23 in the same period). Given these trends, city dwellers tend to spend most of their twenties living with roommates. Finding and keeping a good roommate, however, gets harder as more people head to cities and the market becomes more competitive.

## Case Study: Dance Centres 
Alan Murphy Dance Centres are franchised dance studios in London that are part of a worldwide network of dance studios.<br>
Currently AM Dance Studio is teaching ballroom dance classes to a wide audience of people, who are mostly above 50 but also include younger participants (including few children under 18).
 
At the moment dance studio is struggling to compete with studios operating **within the same area**.<br>
The business model of AM is based on offering students packages, which give access to various individual classes, group classes as well as events.<br>
Ability to purchase **individual classes** is, however, not offered.<br>
As a result prices are higher than average and studio is losing clients to competition.<br>
Furthermore, Alan Murphy tends to attract mostly people with **significant incomes**.<br>
At the moment **younger audiences** choose classes nearby that are either more fitness or fun focused.<br>
In order to continue success story, Alan Murphy needs to **diversify** its client base.
 
Alan Murphy takes pride in the way dancing is taught by the company from the inception of business in 1937.<br>
Studios try to differentiate themselves from competition by attempting to offer high-value service, holistic dance education, unique curriculum, care for students beyond dance class and community of dancers around the world that join together for events and socials.
 
The **quality of dance** teaching is of a highest priority to the studios.<br>
However, at the moment dance teachers are burdened by administrative tasks that are required of them to be performed on daily, weekly and monthly basis.<br>
Each lesson is entered into a computer system manually.<br>
On top of that after each lesson teacher has to mark that lesson as done, which takes additional time (to search for a student, search for a lesson and to mark it).<br>
Data is being stored in a simple spreadsheet fashion however.<br>
That helps teachers to see list of all current students and corresponding lesson schedules as well as all previous students.<br>
However, there is no data analysis currently in place of any sorts.<br>
Organisation has little awareness of what data means in organisational context.
 
Moreover, student information such as dance preferences and specific dance needs are being stored in a **paper format** in folders and teachers have to update those during their class, which further reduces the amount of time that teacher can focus on the lesson itself.<br>
This raises concerns about data privacy as well as folders with sensitive information might be easily seen and possibly accessed by unauthorised persons.
 
All things considered, significant time is spent using outdated system that relies on paper documents, which is also hard to track and analyse in any meaningful way.<br>
Therefore, new digital solution is to be proposed to enhance student care and to improve quality of dance teaching overall as well as security and transparency of the data.
 
### Deliverables
1. Stakeholders
Students should use stakeholder analysis diagram from the lecture to come up with all possible stakeholders and to categorise them appropriately.
2. Brainstorming of ideas
What major pain points can we identify from the brief above? In other words, what are the problematic areas that could be potentially tackled to improve services that the business offers to all its customers? Choose one that team would like to tackle.
This is a discussion of ideas to tackle problem identified above. All ideas are good ideas at this stage, so capture all without judgement. Also assume at this point that we have no business constraints whatsoever. Choose in your groups what you think are your top 5 ideas.
3. Risk Assessment
Whatever ideas have been selected from your brainstorming session, subject them to simple risk assessment as described during lecture. Use risk assessment matrix for that purposes.
4. Detailing of an idea
Based on your risk assessment you should now be able to identify most relevant and feasible idea.
NOTE: This is much more complex in real world but the process is simplified for the sake of time in this exercise. Describe your idea in as much details as possible so that you can present it to the rest of the class at the end of this session. As an output of this step you might have a sketch of your product, a system diagram of the process or anything else that might be relevant and visual.
5. Pitch
This is a pitch of an idea to be delivered by team members to the rest of the class. Treat this as a mini presentation so that you can secure funding for your project. Be snappy and to the point but make sure everyone says few words about the proposed solution and its context (even if it is just a sentence or two).
