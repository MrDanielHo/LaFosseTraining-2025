```python
df['album_name'] = df['album_name'].fillna(df['track_name']+' (single)')
```

```python
# df['featured'] = df['artists'].str.split(';').str[1]
df.dropna(subset=['artists'])
df['artists'] = df['artists'].str.split(';').str[0]
```
