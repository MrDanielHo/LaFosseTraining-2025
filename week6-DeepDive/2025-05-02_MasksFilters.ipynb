## Logical conditions
Python: AND / OR / NOT<br>
Pandas: & / | / !

````python
boy_mask = (df['Sex'] == 'male') & (df['Age'] < 18)
boy_mask.sample(15)
````

````python
boys = df[boy_mask]
boys.head()
````

````python
female_first_class_mask = (df['Sex'] == 'female') & (df['Pclass'] == 1)
female_first_class_mask.sample(15)
````

````python
female_first_class = df[female_first_class_mask]
female_first_class.head(15)
````

````python
older_30_high_fare_mask = (df['Age'] > 30) & (df['Fare'] > 50)
older_30_high_fare = df[older_30_high_fare_mask]
older_30_high_fare.head(10)
````
