## Questions and Hypotheses
1. Which country produces the highest rated ramen?
2. Do the ratings scores vary widely in this dataset?
3. Who are the most and least reviewed brands?
    - Hypothesis: A few popular brands are rated far more than others
4. Which brands most consistently recieve high average ratings?
5. Does the style of ramen (e.g. cup, bowl, pack) affect the ratings?
    - Hypothesis: Ramen served in unusual styles (bottles/cans) will have lower ratings than traditional styles (box) will have higher ratings

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ydata_profiling import ProfileReport

sns.set_style('whitegrid')
```

```python
df = pd.read_csv("./data/ramen.csv", index_col=0)
df.head(10)
```

```python
df.shape
```

```python
# df.isnull.sum()
```

```python
df.drop('T', axis=1, inplace=True)
```

```python
df.info()
```

`Stars` is a rating out of 5.<br>
But the above cell shows us that it is an object/string

```python
df['Stars'].value_counts()
```

We can see Unrated, NR, 4.25/5

```python
no_stars_mask = (df['Stars'] == 'NR') |(df['Stars'] == 'Unrated') | (df['Stars'] == 'NS')
df_no_rating_rows = df[no_stars_mask]
df_no_rating_rows
```

```python
# Tilda
# - removes rows that don't match the conditions of the mask
# - Acts like a NOT operator.
df = df[~no_stars_mask]
df.shape
```

```python
df['Stars'] = df['Stars'].str.replace(r'/.*','',regex=True)
```

```python
df['Stars'].value_counts()
```

```python
df.info()
```

The `Stars` column is still an object/string

```python
df['Stars']  = pd.to_numeric(df['Stars'])
df.info()
```

We check whether the ratings are between 0 and 5 

```python
df = df[(df['Stars'] >= 0) & (df['Stars'] <= 5)]
df.shape
```

```python
df.head()
```

```python
unique_countries_sorted = sorted(df['Country'].unique())
unique_countries_sorted
```

```python
country_avg = df.groupby('Country')['Stars'].mean().sort_values(ascending=False)
country_avg
```

```python
style_avg = df.groupby('Style')['Stars'].mean().sort_values(ascending=False)
style_avg
```

```python
brand_avg = df.groupby('Brand')['Stars'].mean().sort_values(ascending=False)
brand_avg
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=country_avg.index,
            y=country_avg.values)
plt.xticks(rotation=90)
plt.title('Average Ramen Ratings by Country')
plt.xlabel('Country')
plt.ylabel('Average Rating')
plt.show()
```
