## Cleaning the Data


### Company Name

```python
# Remove ratings from the Company name
df['Company Name'] = df['Company Name'].str.split(r'\n', expand=True)[0]
df.head()
```

### Salary


```python
# Fix the Salary Esimate to show min/max values
df['Salary Min'] = df['Salary Estimate'].str.split(r'(', expand=True)[0].str.replace(r'$','').str.replace(r'K','').str.split(r'-', expand=True)[0].astype(int)
df['Salary Max'] = df['Salary Estimate'].str.split(r'(', expand=True)[0].str.replace(r'$','').str.replace(r'K','').str.split(r'-', expand=True)[1].astype(int)
df.head()
```

```python
# Include the mean Salary
df['Salary Avg'] = (df['Salary Min'] + df['Salary Max']) // 2 
df.head()
```

### Ratings

```python
df.sort_values(by='Rating', ascending=True)
```

We have -1.0 ratings where the data is missing.<br>

```python
# Show -1.0 ratings
df[df['Rating'] == -1.0].shape
```

```python
# Replace ratings from -1.0 to 0.0
df['Rating'].replace(to_replace= -1.0, value= 0.0, inplace= True)
df.sort_values(by='Rating', ascending=True)
```

### Job Description

```python
df['Job Description'][0].split('\n')
```

### Languages and Cloud Platforms

```python
df['python'] = df['Job Description'].str.contains('python', case=False)
df['java'] = df['Job Description'].str.contains('java ', case=False)
df['javascript'] = df['Job Description'].str.contains('javascript', case=False)
df['sql'] = df['Job Description'].str.contains('sql', case=False)
df['excel'] = df['Job Description'].str.contains('excel', case=False)
df['aws'] = (df['Job Description'].str.contains('aws', case=False)) | (df['Job Description'].str.contains('amazon', case=False))
df['gcp'] = (df['Job Description'].str.contains('gcp', case=False)) | (df['Job Description'].str.contains('google', case=False))
df['azure'] = (df['Job Description'].str.contains('azure', case=False)) | (df['Job Description'].str.contains('microsoft', case=False))
df.head()
```

### Job Title

```python
df['Job Title'].value_counts()
```

```python
# Function for job titles stating 'data scientist' to return 'data scientist' 
# above everything else and to follow the rest of the if statement.

def job_title_classification(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning engineer'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'n/a'


# Function for job titles mentioning their senority to be indicated.
def seniority(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'sr' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
            return 'senior'
    elif 'jr' in title.lower() or 'jr.' in title.lower():
        return 'junior'
    else:
        return 'n/a'
```

```python
df['Job Simplified'] = df['Job Title'].apply(job_title_classification)
df['Job Simplified'].value_counts()
```

```python
df['Senority'] = df['Job Title'].apply(seniority)
df['Senority'].value_counts()
```

### Drop

```python
# Drop columns not used in our questions
df.drop(['Competitors', 'Founded'], axis=1, inplace=True)
df.head()
```

```python
# Apply str.strip() to the entire DataFrame
df = df.map(lambda x: x.strip() if type(x) == str else x)
```
